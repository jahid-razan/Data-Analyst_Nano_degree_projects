### 3.4.2 Define Tidyness issues in df_twitter dataframe

### Tidyness Issue 1

**Define: The datetime format is difficult to read. Extract Year, Month, and Day and hour from it.**

**Code**

```python
# finding the year from the start date
df_twitter['year'] = df_twitter['timestamp'].dt.year

# Extracting Day from datetime
df_twitter['day'] = df_twitter['timestamp'].dt.strftime('%a')

# Creating the month category- will return the month of the year
df_twitter['month'] = df_twitter['timestamp'].dt.strftime('%b')

#extracting hour- will return a value from 0 to 23 
df_twitter['hour'] = df_twitter['timestamp'].dt.hour.astype(int)
```

### Tidyness Issue 2

** Define: Bring every dog stage information to rows and get rid of the stage columns.**

**Code**

```python
#combining all the dog stages-doggo, floofer, pupper, puppo into stages column
df_twitter['stages']= df_twitter[['doggo', 'floofer','pupper','puppo']].fillna('').sum(axis=1)
```

**Note: It seems that some stages has been combined. Considering a None value for the abscence of stage value we expect 05 unique stage data- which are- None, doggo, floofer, pupper, puppo. Hence we need to exclude the rows where stage data has been combined.**

```python
#find where doggo and puppo has been combined
df_twitter.query('doggo=="doggo" & puppo == "puppo"')
```
